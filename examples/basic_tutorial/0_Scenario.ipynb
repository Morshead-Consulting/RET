# RET Basic Tutorial

## Tutorial Scenario and Hypotheses

The scenario being developed in this tutorial is deliberately basic and has been designed solely to support learning to use RET.  The scenario is built to exercise key aspects of RET - agents, models, data collectors and batch runners.

## Overview

For the purposes of this tutorial we will construct a simple assault onto a defended position.  Red forces will defend a key location using anti-armour, infantry and air defence capabilities.  Blue will attempt to destroy the Red air defence capabilities with a pre-planned fire mission, observe into the defenced position using unmanned air vehicles (UAVs), and then assault the position with the intent of driving Red forces back and seizing the key ground.

## Plans

### Blue

At H0 the Rocket Artillery will launch a pre-planned attack on the perceived location of the Red Air Defence (AD) units.  Following completion of that mission a single UAV will be tasked with over flying the area to perform Battle Damage Assessment (BDA).  If the AD units are observed to be still alive or if the UAV is shot down then a second fire mission willbe fired into the area.  Once completed the remaining UAVs will fly forward to observe the approach to the defended position and into the position itself and the Armoured Infantry (AI) units will begin their assault.  If any unit observes the Anti-armour units then the rocket artillery will be tasked with firing onto their position.  The AI units will attack any unit they identify.

In the event that more than 3 sections of AI have been destroyed, the Blue forces will withdraw to the start location and the mission will be deemed a failure.

Blue will claim the mission is a success when all Red units in the defended position have been destroyed.

### Red

All Red forces will have standing orders to engage appropriate Blue targets when they are recognised.  The Anti-armour units are waiting in the dense forest hoping to ambush advancing Blue forces.

Red will withdraw from the defended position if more than 1 mechanised infantry section is destroyed.  This will be deemed a mission failure.

Red will claim the mission is a success in the event that all Blue units are destroyed or if Blue withdraws to their start position.
